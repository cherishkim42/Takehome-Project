```python
from scipy import stats
import math
import csv
import re
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = pd.read_csv('seattle-crime-stats-by-1990-census-tract-1996-2007.csv')
df.head()
```

### Let's see the different values present in each of the three descriptive columns.

```python
df['Crime_Type'].unique()
```

```python
df['Report_Year'].unique()
```

```python
df['Census_Tract_1990'].unique()
```

### Let's narrow it down a bit; I'll look at the data for Census Tract 1.

```python
census_1 = df[(df['Census_Tract_1990'] == 1.)]
census_1.head()
```

### Will we observe notable trends in the number of aggravated assaults by year?

```python
census_1_aa = census_1[(census_1['Crime_Type'] == 'Aggravated Assault')]
census_1_aa.head()
```

```python
census_1_aa.plot.line(x='Report_Year', y='Report_Year_Total')
```

### ... yes, but not by much.

### Now let's get a visual representation of the different types of crimes.

```python
census_1_crime = census_1.sort_values(by='Crime_Type')
ax = sns.boxplot(x='Crime_Type', y='Report_Year_Total', data=census_1_crime)
```

### How does the breakdown of different crimes look for the year 2007?

```python
census_1_2007 = census_1[(census_1['Report_Year']==2007)]
census_1_2007
```

```python
plt.pie(census_1_2007['Report_Year_Total'], labels=['Aggravated Assault', 'Homicide', 'NonResidential Burglary', 'Rape', 'Residential Burglary', 'Robbery', 'Theft', 'Vehicle Theft'], autopct='%1.1f%%', shadow=True)
center_circle = plt.Circle((0,0),0.75,color='black', fc='white', linewidth=1.25)

fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.axis('equal')
```

### Throwback time! Let's do the same, but for the year 1996.

```python
census_1_1996 = census_1[(census_1['Report_Year']==1996)]
census_1_1996
```

```python
plt.pie(census_1_1996['Report_Year_Total'], labels=['Aggravated Assault', 'Homicide', 'NonResidential Burglary', 'Property Crimes Total', 'Rape', 'Residential Burglary', 'Robbery', 'Theft', 'Vehicle Theft', 'Violent Crimes Total'], autopct='%1.1f%%', shadow=True)
center_circle = plt.Circle((0,0),0.75,color='black', fc='white', linewidth=1.25)

fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.axis('equal')
```
